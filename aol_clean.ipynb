```python
import ir_datasets
import pandas as pd
from datetime import datetime, timedelta
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
aol = ir_datasets.load("aol-ia")
```

```python
def parse_time(ts):
    return datetime.strptime(ts, "%Y-%m-%d %H:%M:%S")
```

```python
train = []
val = []
test = []

train_len = 0
val_len = 0
test_len = 0
for rec in aol.qlogs_iter():
    dic = {
        "user_id": rec.user_id,
        "query_id": rec.query_id,
        "query": rec.query,
        "query_orig": rec.query_orig,
        "time": rec.time,
    }
    
    if rec.time.date() <= datetime(2006, 3, 5).date():
        train.append(dic)
        train_len += 1

        if train_len % 1000 == 0:
            print(f"train: {train_len}")
    elif rec.time.date() == datetime(2006, 3, 6).date():
        val.append(dic)

        val_len += 1
        if val_len % 1000 == 0:
            print(f"val: {val_len}")
    elif rec.time.date() == datetime(2006, 4, 20).date():
        test.append(dic)

        test_len += 1
        if test_len % 1000 == 0:
            print(f"test: {test_len}")
```

```python
train_df = pd.DataFrame(train)
val_df = pd.DataFrame(val)
test_df = pd.DataFrame(test)
```

```python
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)
```
